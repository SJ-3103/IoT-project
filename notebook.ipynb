```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('my_database.csv')
```

```python
df.head(10)
```

```python
plt.figure(figsize=(10, 5), layout='constrained')
plt.plot(df['DateTime'], df['Temperature'], label='linear')
plt.xlabel('Datetime')
plt.ylabel('Temperature')
plt.title("Temerature vs Datetime")
plt.xticks(rotation=90)
plt.legend();
```

```python
plt.figure(figsize=(10, 5), layout='constrained')
plt.plot(df['DateTime'], df['Humidity'], label='linear')
plt.xlabel('Datetime')
plt.ylabel('Humidity')
plt.title("Humidity vs Datetime")
plt.xticks(rotation=90)
plt.legend();
```

```python
plt.figure(figsize=(10, 5), layout='constrained')
plt.plot(df['DateTime'], df['LightVal'], label='linear')
plt.xlabel('Datetime')
plt.ylabel('Light Value')
plt.title("LightVal vs Datetime")
plt.xticks(rotation=90)
plt.legend();
```

```python
plt.figure(figsize=(10, 5), layout='constrained')
plt.plot(df['DateTime'], df['MoistureVal'], label='linear')
plt.xlabel('Datetime')
plt.ylabel('Moisture')
plt.title("Moisture vs Datetime")
plt.xticks(rotation=90)
plt.legend();
```

```python
plt.figure(figsize=(10, 5), layout='constrained')
plt.plot(df['DateTime'], df['Temperature'], label='linear')
plt.plot(df['DateTime'], df['Humidity'], label='linear')
plt.plot(df['DateTime'], df['LightVal'], label='linear')
plt.plot(df['DateTime'], df['MoistureVal'], label='linear')
plt.xlabel('Datetime')
plt.title("Temerature,Humidity,LightVal,Moisture vs Datetime")
plt.xticks(rotation=90)
plt.legend();
```
